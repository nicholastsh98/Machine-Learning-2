```python
import time

import torch
from torch import Tensor
torch.manual_seed(0)
from torch import nn
import torch.nn.functional as F
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
stdscaler = StandardScaler()

def closure():
        optimizer.zero_grad()
        y_hat = model.forward(X_train)
        loss = criterion(y_hat, y_train)
        loss.backward()
        return loss
    
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()

import numpy as np
from sklearn.model_selection import GridSearchCV

from copy import deepcopy

data = pd.read_csv('Proj1ReorderedRound.csv')

import warnings
warnings.filterwarnings('ignore')
```

```python
X = data.drop('Condition', axis=1).values
y = data['Condition'].values

y = LE.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train = stdscaler.fit_transform(X_train)
X_test = stdscaler.transform(X_test)
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)
```

```python
class ANN(nn.Module):
    def __init__(self,solver, array):
        super().__init__()
        
        if (solver == 'SGD'):
            self.fc1 = nn.Linear(in_features = 12, out_features = array[0])
            self.output = nn.Linear(in_features=array[0], out_features=3)
            
        if (solver == 'LBFGS'):
            self.fc1 = nn.Linear(in_features = 12, out_features = array[0])
            self.fc2 = nn.Linear(in_features = array[0], out_features = array[1])
            self.output = nn.Linear(in_features=array[1], out_features=3)
        
        if (solver == 'Adam'):
            self.fc1 = nn.Linear(in_features = 12, out_features = array[0])
            self.fc2 = nn.Linear(in_features = array[0], out_features = array[1])
            self.fc3 = nn.Linear(in_features = array[1], out_features = array[2])
            self.output = nn.Linear(in_features=array[2], out_features=3)
            
        self.layers = len(array)
        
        self.solver = solver
 
    def forward(self, x):
        
        if (solver == 'SGD'):
            x = F.selu(self.fc1(x))
            x = self.output(x)
            
        if (solver == 'LBFGS'):
            x = F.relu(self.fc1(x))
            x = F.relu(self.fc2(x))
            x = self.output(x)
        
        if (solver == 'Adam'):
            x = F.selu(self.fc1(x))
            x = F.selu(self.fc2(x))
            x = F.selu(self.fc3(x))
            x = self.output(x)
        
        return x
    
alpha = 0.1
criterion = nn.CrossEntropyLoss()
```

```python
initiator = 0

solver = 'Adam'         #choose solver here

if (solver == 'Adam'):
    af = 'selu'
    layers = 3

if (solver == 'LBFGS'):
    af = 'relu'
    layers = 2
    
if (solver == 'SGD'):
    af = 'selu'
    layers = 1


basename = ['i','j','k']
    
configDF = pd.DataFrame(columns = basename[:layers])
index = 0

for i in range(12):
    if (layers == 1):
        configDF.loc[index] = [i+1]
        
    for j in range(12):
        if (layers == 2):
            configDF.loc[index] = [i+1,j+1]
            
        for k in range(12):
            if (layers == 3):
                configDF.loc[index] = [i+1,j+1,k+1]
                
            index += 1
            
configDF.index = range(len(configDF))





print('Solver : ', solver)
print('Activation function : ', af)
print('# of hidden layers : ', layers)

configDF
```

```python
for index in configDF.index:
    Arr = configDF.loc[index].values
    print(Arr)
```

```python
time_list = []

epochs = 100

LossDF = pd.DataFrame(columns = ['config','nodes','initial','midway','final','last change'])

if (initiator == 0):    #checks if this cell has been run before...
    FML = []   # Full Model List : stores initial model weights and bias for future reference
      
for index in configDF.index:
    Start = time.time()     #start time
    to_add = []
    
    Arr = configDF.loc[index].values
    to_add.append(Arr)
    to_add.append(configDF.loc[index].sum())
    
    if (initiator == 0):        # if this cell wasnt run before, store the starting weights and bias for future ref
        model = ANN(solver,Arr)
        dummy = deepcopy(model)
        FML.append(dummy)
        
    elif (initiator == 1):      # otherwise, use the same starting params...
        model = deepcopy(FML[index])
        setattr(model, 'array', Arr)    #... while overwriting the hidden layer configuration
        
    if (solver == 'Adam'):
        optimizer = torch.optim.Adam(model.parameters(), lr=alpha) # ADAM
    if (solver == 'LBFGS'):
        optimizer = torch.optim.LBFGS(model.parameters(), lr=alpha) # LBFGS
    if (solver == 'SGD'):
        optimizer = torch.optim.SGD(model.parameters(), lr=alpha) # SGD
    
    for e in range(epochs):
        y_hat = model.forward(X_train)
        loss = criterion(y_hat, y_train)
                
        if e == 0:
            to_add.append(loss.detach().numpy())
                    
        if e == 49:
            to_add.append(loss.detach().numpy())
 
        optimizer.zero_grad()
        loss.backward()
             
        if (solver == 'LBFGS'):
            optimizer.step(closure)
            
        else:
            optimizer.step()
        
        if e == 98:
            A = loss.detach().numpy()
            
        if e == 99:
            B = loss.detach().numpy()
            
    to_add.append(loss.detach().numpy())
    to_add.append(B-A)
            
    LossDF.loc[index] = to_add
    
    End = time.time()
    dura = End - Start
    time_list.append(dura)
    
    print('\nComputation for',Arr,'complete')
    print('Time taken : ', np.round(dura,3),'s')

initiator = 1

print('\nSolver : ', solver)
print('# of hidden layers :', layers)
print('Activation function : ', af)
print('Epochs : ', epochs)

print('\nTotal time taken :',np.round(np.sum(time_list),3),'s')

print('average loss = ', np.round(LossDF.mean()[-2],5))
print('average loss (without ones) = ', np.round(LossDF.drop(0).mean()[-2],5))

LossDF
```
